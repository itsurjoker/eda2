```python
#Q2.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df=pd.read_excel('flight_price.xlsx')


flight_prices = df['Price']

plt.figure(figsize=(10, 6))
plt.hist(flight_prices, bins=30, color='skyblue', edgecolor='black')
plt.title('Flight Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q3.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df=pd.read_excel('flight_price.xlsx')


flight_prices = df['Price']

min_price = flight_prices.min()
max_price = flight_prices.max()


price_range = max_price - min_price

print(f"The minimum price is: {min_price}")
print(f"The maximum price is: {max_price}")
print(f"The range of prices is: {price_range}")
```

```python
#Q4.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


df=pd.read_excel('flight_price.xlsx')
airline_column = 'Airline'
price_column = 'Price'

# Create a boxplot to compare prices by airline
plt.figure(figsize=(12, 6))
sns.boxplot(x=airline_column, y=price_column, data=df, palette="Set3")
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q13.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')
df['Price'] = df['Price'].apply(lambda x: float(x.replace('$', '')) if '$' in str(x) else 0)


average_prices = df.groupby('Type')['Price'].mean()


plt.figure(figsize=(8, 6))
average_prices.plot(kind='bar', color='skyblue')
plt.title('Average Prices by App Type (Free vs. Paid)')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.show()
```

```python
#Q14.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

df=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

installations_freq = df['Installs'].value_counts().reset_index()


installations_freq.columns = ['Installs', 'App Count']

# Sort the frequency table by the number of installations in descending order
installations_freq = installations_freq.sort_values(by='App Count', ascending=False)

# Display the top 10 most popular apps
top_10_apps = installations_freq.head(10)
print(top_10_apps)
```
